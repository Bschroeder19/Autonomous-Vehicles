```python
##Applying arbitrary edge detection kernels to an image to aid the identification of lane lines
```

```python
kernel_identity = np.array([[0, 0, 0],
                            [0, 1, 0],
                            [0, 0, 0]])
```

```python
sobel_y_kernel = np.array([[1, 2, 1],
                           [0, 0, 0],
                           [-1, -2, -1]])
```

```python
sobel_y_image = cv2.filter2D(img, -1, sobel_y_kernel)
```

```python
show_with_matplotlib(sobel_y_image, "sobel y image", 200)
```

```python
##Overlaying text to describe the use of the kernel
```

```python
colors = {'blue': (255, 0, 0), 'green': (0, 255, 0), 'red': (0, 0, 255), 'yellow': (0, 255, 255),
          'magenta': (255, 0, 255), 'cyan': (255, 255, 0), 'white': (255, 255, 255), 'black': (0, 0, 0),
          'gray': (125, 125, 125), 'rand': np.random.randint(0, high=256, size=(3,)).tolist(),
          'dark_gray': (50, 50, 50), 'light_gray': (220, 220, 220)}
```

```python
image = np.zeros((400, 1200, 3), dtype="uint8")
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1.5
thickness = 5
text = 'Sobel kernels-aid the detection of lane lines'
circle_radius = 10
```

```python
ret, baseline = cv2.getTextSize(text, font, font_scale, thickness)
```

```python
text_width, text_height = ret
```

```python
text_x = int(round((image.shape[1] - text_width) / 2))
text_y = int(round((image.shape[0] + text_height) / 2))
```

```python
cv2.circle(sobel_y_image, (text_x, text_y), circle_radius, colors['black'], -1)
```

```python
cv2.putText(sobel_y_image, text, (text_x, text_y), font, font_scale, colors['blue'], thickness)
```
